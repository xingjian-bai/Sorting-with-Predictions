```python
import random
def random_suffix():
    suffix = ''
    for _ in range(3):
        suffix += str(random.randint(0, 9))
    return suffix
```

```python
import matplotlib.pyplot as plt
import math
def plot_data(names, data, n, plot_title, exclude = [], target = [], plot_name='plot', scale='log'):
    x = range(len(data[0]))
    if scale == 'linear':
        x = [2**i for i in x]
    print(x)
    for (i, row) in enumerate(data):
        y = [i / n for i in row]
        if names[i] in exclude:
            continue
        if names[i] in target:
            # make the first two lines wider
            plt.plot(x, y, label=names[i], linewidth=3)
        else:
            plt.plot(x, y, label=names[i])
    
    nlogn_y = [math.log2(n) for i in x]

    plt.plot(x, nlogn_y, 'r--', label='nlogn')

    # x_ticks = [f"{2**i / n:.1f}" for i in x]
    # plt.xticks(x, x_ticks)

    plt.xlabel(f'error ({scale} scale)')
    plt.ylabel('# comparisons')
    plt.title(plot_title)
    plt.legend()
    plt.savefig('plots/' + plot_name + random_suffix() + '.png')
    plt.show()
```

```python
# load txt 
def read_data(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
    
    # first line is names
    print(lines[0])
    names = lines[0].strip().split(' ')
    lines = lines[1:]

    data = []
    for line in lines:
        row = line.strip().split(' ')
        row_data = [int(num) for num in row if num]
        if row_data:
            data.append(row_data)

    return names, data
```

```python
def experiment (file_name, n, plot_name, exclude = [], target = [], plot_title = 'plot', scale = 'log'):
    names, data = read_data(file_name)
    for (i, datum) in enumerate(data):
        print(names[i], datum)
    plot_data(names, data, n, plot_name, exclude, target, plot_title, scale)
    # generate random suffix
    
```
